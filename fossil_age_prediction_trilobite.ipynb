```python
import pandas as pd

data = pd.read_csv("trilobite_occurrences.csv")
```

```python
print(data.isnull().sum())
```

```python
data["genus"].fillna("Unknown",inplace=True)
data["formation"].fillna("Unknown",inplace=True)
data["environment"].fillna("Unknown",inplace=True)
```

```python
data.dtypes
```

```python
x = data.drop(["occurrence_no","identified_name","identified_rank","genus","species_name","formation","environment","time_bins","max_ma","min_ma","mid_ma"],axis=1)
y = data["mid_ma"]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(
    x,y,test_size=0.2,random_state = 20
)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()

model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
print("Predictions :",y_pred)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print("MAE :",mae)
print("MSE :",mse)
print("R2 :",r2)
```

```python
print("Actual values :",y_test)
print("Predicted values :",y_pred)
```
